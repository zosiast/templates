```python
import cf
import cfplot as cfp
```

## Read in data using cf

```python
xl_ch4 =cf.read('/shared/netscratch/ptg21/UM_data/xltln/xltln_ch4.nc')[0]
xl_ch4_ppb = xl_ch4*28/16.*1e9 #unit conversion to ppb
xl_ch4_1 = xl_ch4_ppb[:,0,:,:] #first model level ony
```

## Contour plots

```python
#cfp.levs(min=1600, max=2800, step = 100) #set colorbar levels
cfp.con(xl_ch4_ppb[0,0,:,:], lines=False, title='Surface CH4 for first timestep') #set time and model level
```

```python
#cfp.levs(min=1600, max=2800, step = 100) #set colorbar levels
cfp.con(xl_ch4_ppb[0,4,:,:], lines=False, title='CH4 for first timestep, model level 5') #set time and model level
```

```python
cfp.con(xl_ch4_1.collapse('mean','time'), lines=False, title='Annual mean surface CH4 xltln')
```
