```python
import matplotlib
import numpy as np
import numpy.ma as ma
import pylab as plt
```

```python
#create random array
ch4_diff_non_mask = np.random.randint(1,100, size=(144,30))
ch4_diff = ma.masked_where(ch4_diff_non_mask < 40, ch4_diff_non_mask)

x_vals = np.array(list(range(144)))
y_vals = np.array(list(range(30)))
```

```python
print(ch4_diff_non_mask.shape)
```

```python
#test stipple mask
plt.figure(dpi=150)

miss_vals = ma.masked_where(ch4_diff.mask == False,np.ones((144,30)))

#change vmin and vmax to change intensity of colour of stippling
miss = plt.pcolor(miss_vals,cmap='Greys', vmin=0,vmax=9, hatch='....', edgecolor='lightgrey')

plt.xlabel('Time')
plt.ylabel('Latitude')
plt.title(f'Stippled out data')
```
