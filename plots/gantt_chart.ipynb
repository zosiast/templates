```python
fig = plt.figure(dpi=150,figsize=(12,5))
ax = plt.axes()

y_tick_labs = ['Writing up','Explore removal scenarios',
               'Data analysis',
               'Experiments: Table 3','Experiments: Table 2','Continuing model evaluation']
x_tick_labs = ['Jul 20','Jul 21','Jul 22','Jul 23']

ax.set_ylim(0, 35)
ax.set_xlim(0, 36)

ax.set_yticks([5,10,15,20,25,30]) 
ax.set_yticklabels(y_tick_labs)

ax.set_xticks([0,12, 24, 36]) 
ax.set_xticklabels(x_tick_labs)
ax.grid(axis='x', which = 'major',linestyle='--')

plt.rc('xtick', labelsize=20)    # fontsize of the tick labels
plt.rc('ytick', labelsize=20) 

#define time range of tasks
ax.broken_barh([(0, 3)], (28, 4),facecolors='lightblue')
ax.broken_barh([(3, 9)], (23, 4),facecolors='lightblue')
ax.broken_barh([(8, 4)], (18, 4),facecolors='lightblue')
ax.broken_barh([(6, 12)], (13, 4),facecolors='lightblue') 
ax.broken_barh([(12, 10)], (8, 4),facecolors='lightblue')
ax.broken_barh([(20, 12)], (3, 4),facecolors='lightblue')
 
#ax.set_xlabel('Time') 
#ax.set_ylabel('Task')
#plt.savefig("gantt1.png") 
```
