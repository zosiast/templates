```python
import netCDF4 as nc
import pylab as plt
import numpy as np
from mpl_toolkits.basemap import Basemap, shiftgrid
import matplotlib
```

```python
data = nc.Dataset('/shared/netscratch/ptg21/UM_data/xltln/xltln_ch4.nc')
#extract variables to arrays
ch4 = data.variables['ch4'][:]*28/16.*1e9
lat = data.variables['latitude'][:]
lon = data.variables['longitude'][:]
alt = data.variables['level_height'][:] #alt in m

ch4, lon = shiftgrid(180.,ch4,lon,start=False)
```

## One latitude

```python
fig = plt.figure(dpi=100)
ax = plt.axes()

ind = list(lat).index(60) #select lat value
test = ch4[0,:,ind,0]
ax.plot(test,alt)

plt.xlabel('[CH4] / ppb')
plt.ylabel('Model height')
plt.title(f'CH4 Profile at lat = {ind}')
```

## Selected latitudes

```python
ch4_mean = np.mean(ch4, axis=(0,3)) #mean over lon and time

fig = plt.figure(dpi=100)
ax = plt.axes()

for i in range(-90,90,30):
    ind = list(lat).index(i)
    lat_data = ch4_mean[:,ind]
    ax.plot(lat_data,alt, label=f'{i}')
    
#ind = list(lat).index(-60) #select lat value
#test = ch4[0,:,ind,0]
#ax.plot(test,alt)
plt.legend()
plt.xlabel('[CH4] / ppb')
plt.ylabel('Model height')
plt.title(f'CH4 Profile at different latitudes')
```

## Latitude bins

```python
fig = plt.figure(dpi=150)
ax = plt.axes()
cmap = matplotlib.cm.get_cmap('viridis')

base=np.zeros(6)

for i in range(-90,90,30):
    ind1 = list(lat).index(i)
    ind2 = list(lat).index(i+30)
    lat_data = ch4_mean[:,ind1:ind2]
    lat_mean = np.mean(lat_data, axis=1)
    #print(lat_mean[0])
    ind3=np.int((i+90)/30)
    base[ind3] = lat_mean[0]
    ax.plot(lat_mean - base[0], alt, label=f'{i} to {i+30}',c=cmap((i+90)/150))
    
plt.legend(title='Latitude')
ax.set_ylim(0,12000)
ax.set_xlim(-250,250)
plt.xlabel('[CH4] / ppb difference from -90_-60 surface')
plt.ylabel('Altitude / m')
plt.title(f'CH4 Profile: model with emissions (xltln)')
```
