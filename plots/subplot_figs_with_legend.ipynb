```python
import pylab as plt
import numpy as np
```

```python
#data
yvals = np.linspace(0,100)
xvals = np.linspace(0,50)
```

```python
fig = plt.figure(dpi=300)

plt_labels=["-90_-60","-60_-30","-30_0","0_30","30_60","60_90"]

for i in range(6):
    ax = plt.subplot(2,3,i+1)
    #plot data
    ax.plot(xvals,yvals,c='k',label=f'obs')
    ax.plot(xvals+5,yvals,label=f'series2')
    ax.plot(xvals-5,yvals,label=f'series3')
    
    ax.set_ylim(0,100)
    ax.set_xlim(0,50)
    
    plt.xlabel('[CH4] / ppb diff')
    plt.ylabel('Altitude / km')
    plt.title(f'{plt_labels[i]}')
    
    #location of legend
    if i == 5:
        plt.legend(bbox_to_anchor=(1.8,1.4))

#only label outer axes
for ax in fig.get_axes():
    ax.label_outer()

#for legend for all lines in all subplots uncomment next line
#fig.legend(bbox_to_anchor=(1.1,1))
```
